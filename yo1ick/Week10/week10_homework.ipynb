```python
import os
from dashscope import MultiModalConversation
import dashscope 


local_path = "/Users/yolick/Desktop/Week10_副本/WechatIMG25.jpg"
image_path = f"file://{local_path}"
messages = [
                {'role':'user',
                'content': [{'image': image_path},
                            {'text': '请判断这张图片中是狗(dog)还是猫(cat)，只输出分类结果'}]}]
response = MultiModalConversation.call(
    api_key='sk-3dc5716884f046e286349aaa6da865e4',
    model='qwen3-vl-plus',  # 此处以qwen3-vl-plus为例，可按需更换模型名称。模型列表：https://help.aliyun.com/zh/model-studio/models
    messages=messages)
print(response["output"]["choices"][0]["message"].content[0]["text"])
```
